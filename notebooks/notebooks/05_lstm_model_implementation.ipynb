```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

import torch
import torch.nn as nn

from torch.utils.data import Dataset, DataLoader

import pytorch_lightning as pl

import warnings
warnings.filterwarnings("ignore")
```

#### Load Dataset

```python
df = pd.read_csv("../../data/cleaned/time_series_data.csv")
```

```python
df.head()
```

#### Extract Target Variable

```python
sales = df["y"].values.reshape(-1, 1)
```

#### Scale data

```python
scaler = MinMaxScaler()

sales_scaled = scaler.fit_transform(sales)
```

#### Create Sequences

```python
def create_sequences(data, seq_length=30):

    X = []
    y = []

    for i in range(len(data) - seq_length):

        X.append(data[i:i+seq_length])

        y.append(data[i+seq_length])

    return np.array(X), np.array(y)
```

```python
X, y = create_sequences(
    sales_scaled,
    seq_length=30
)

print(X.shape)
print(y.shape)
```

#### Train-Test split

```python
split = int(len(X) * 0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

#### Dataset class

```python
class SalesDataset(Dataset):

    def __init__(self, X, y):

        self.X = torch.tensor(
            X,
            dtype=torch.float32
        )

        self.y = torch.tensor(
            y,
            dtype=torch.float32
        )

    def __len__(self):

        return len(self.X)

    def __getitem__(self, idx):

        return self.X[idx], self.y[idx]
```

#### dataLoaders

```python
train_loader = DataLoader(
    SalesDataset(X_train, y_train),
    batch_size=32,
    shuffle=True
)

test_loader = DataLoader(
    SalesDataset(X_test, y_test),
    batch_size=32
)
```

#### LSTM Model

```python
class LSTMForecast(pl.LightningModule):

    def __init__(self):

        super().__init__()

        self.lstm = nn.LSTM(
            input_size=1,
            hidden_size=64,
            num_layers=2,
            batch_first=True
        )

        self.fc = nn.Linear(
            64,
            1
        )

        self.loss_fn = nn.MSELoss()

    def forward(self, x):
        output, _ = self.lstm(x)
        output = output[:, -1, :]
        return self.fc(output)
    
    def training_step(
        self,
        batch,
        batch_idx
    ):

        X, y = batch

        pred = self(X)

        loss = self.loss_fn(
            pred,
            y
        )

        self.log(
            "train_loss",
            loss
        )

        return loss
    
    def configure_optimizers(self):

        return torch.optim.Adam(
            self.parameters(),
            lr=0.001
        )    

```

#### Train Model

```python
model = LSTMForecast()
```

```python
trainer = pl.Trainer(
    max_epochs=20,
    enable_checkpointing=False,
    logger=False
)
```

```python
trainer.fit(
    model,
    train_loader
)
```

#### Predictions

```python
model.eval()
predictions = []
with torch.no_grad():

    for X_batch, _ in test_loader:

        pred = model(X_batch)

        predictions.extend(
            pred.numpy()
        )
```

#### Convert Back to Original Scale

```python
predictions = scaler.inverse_transform(
    np.array(predictions)
)
actual = scaler.inverse_transform(
    y_test
)
```

#### Plot Results

```python
plt.figure(figsize=(12,5))

plt.plot(
    actual,
    label="Actual"
)

plt.plot(
    predictions,
    label="Predicted"
)

plt.title(
    "LSTM Forecast Results"
)

plt.legend()

plt.show()
```

#### Evaluation

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(
    mean_squared_error(
        actual,
        predictions
    )
)

print("RMSE:", rmse)
```
